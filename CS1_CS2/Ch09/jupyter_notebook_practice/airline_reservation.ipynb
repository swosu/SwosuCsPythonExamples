# Goal: Define the Seat Class

In this cell, we are creating a `Seat` class that represents a single seat on an airplane.

The goals of this class are:

- Store identifying information about a seat (seat number, class type)
- Track whether the seat is reserved
- Associate a passenger with the seat if reserved
- Enforce rules about double-booking
- Demonstrate encapsulation and private attributes
- Demonstrate properties
- Demonstrate object behavior via methods

This class is the foundational building block of the reservation system.

```python
class Seat:
    """
    Represents a single seat on an airplane.
    """

    def __init__(self, seat_number: str, seat_class: str):
        self.seat_number = seat_number          # e.g., "12A"
        self.seat_class = seat_class            # Economy, Business, First
        self._is_reserved = False               # private state
        self._passenger = None

    def reserve(self, passenger):
        if self._is_reserved:
            raise ValueError(f"Seat {self.seat_number} is already reserved.")
        self._is_reserved = True
        self._passenger = passenger

    def cancel_reservation(self):
        self._is_reserved = False
        self._passenger = None

    @property
    def is_reserved(self):
        return self._is_reserved

    def __str__(self):
        status = "Reserved" if self._is_reserved else "Available"
        return f"Seat {self.seat_number} ({self.seat_class}) - {status}"
```

# Goal: Define the Passenger Class

In this cell, we create a `Passenger` class.

The goals are:

- Store personal identifying information
- Demonstrate simple object modeling
- Provide behavior through methods
- Show how objects represent real-world entities

This class represents a real person interacting with the airline system.

```python
class Passenger:
    """
    Represents a passenger booking a flight.
    """

    def __init__(self, first_name: str, last_name: str, passport_number: str):
        self.first_name = first_name
        self.last_name = last_name
        self.passport_number = passport_number

    def full_name(self):
        return f"{self.first_name} {self.last_name}"

    def __str__(self):
        return f"{self.full_name()} (Passport: {self.passport_number})"
```

    Objects encapsulate state.

    A behavior method.
    Combines stored data into meaningful output.

__str__:
    Improves usability.
    When printed, the object becomes human-readable.

# Goal: Define the Flight Class (Composition)

In this cell, we define a `Flight` class.

This class demonstrates:

- Composition (a flight contains seats)
- Dictionary storage of objects
- Business logic enforcement
- Separation of responsibilities

The Flight manages seat reservation behavior.

```python
class Flight:
    """
    Represents a specific flight with seats.
    """

    def __init__(self, flight_number: str, origin: str, destination: str):
        self.flight_number = flight_number
        self.origin = origin
        self.destination = destination
        self.seats = {}

    def add_seat(self, seat: Seat):
        self.seats[seat.seat_number] = seat

    def reserve_seat(self, seat_number: str, passenger: Passenger):
        if seat_number not in self.seats:
            raise ValueError("Invalid seat number.")
        self.seats[seat_number].reserve(passenger)

    def available_seats(self):
        return [seat for seat in self.seats.values() if not seat.is_reserved]

    def __str__(self):
        return f"Flight {self.flight_number} from {self.origin} to {self.destination}"
```

# Discussion: Flight Class and Composition

self.seats = {}
    Dictionary used for fast lookup.
    Keys are seat numbers.
    Values are Seat objects.

add_seat:
    Stores a Seat object.
    Demonstrates object aggregation.

reserve_seat:
    Delegates responsibility to Seat.
    Flight does not manage reservation details directly.
    This separation improves maintainability.

available_seats:
    Demonstrates list comprehension.
    Filters based on object state.

This class shows composition:
    A Flight HAS Seats.

# Goal: Define the Airline Class

This class represents the top-level manager.

It demonstrates:

- Aggregation of flights
- Object coordination
- Organizational hierarchy
- Responsibility delegation

```python
class Airline:
    """
    Represents an airline company managing flights.
    """

    def __init__(self, name: str):
        self.name = name
        self.flights = {}

    def add_flight(self, flight: Flight):
        self.flights[flight.flight_number] = flight

    def find_flight(self, flight_number: str):
        return self.flights.get(flight_number)

    def __str__(self):
        return f"Airline: {self.name}"
```

# Discussion: Airline Class

self.flights = {}
    Stores Flight objects.

add_flight:
    Aggregates flight objects.

find_flight:
    Uses dictionary lookup.
    .get() prevents KeyError.

This class demonstrates system hierarchy:

Airline
  → Flight
      → Seat

# Goal: Demonstrate the System in Action

In this cell we:

- Create objects
- Link objects together
- Demonstrate reservation flow
- Print system state

This shows object collaboration.

```python
# Create airline
airline = Airline("Python Airways")

# Create flight
flight = Flight("PA101", "Oklahoma City", "Denver")

# Add seats
for row in range(1, 6):
    for letter in ["A", "B", "C"]:
        seat_number = f"{row}{letter}"
        flight.add_seat(Seat(seat_number, "Economy"))

airline.add_flight(flight)

# Create passenger
passenger = Passenger("Jeremy", "Evert", "X1234567")

# Reserve seat
flight.reserve_seat("1A", passenger)

# Check seats
for seat in flight.seats.values():
    print(seat)
```
