## Section 1: The Problem

Here is the challenge we are going to solve:

> Write a function called `draw_triangle()` that prints an **upside-down isosceles triangle** made of `*` characters.
> The function takes one integer parameter: the **base length** of the triangle.
> The base is always odd and less than 20.
> Each row has **2 fewer stars** than the previous row, and is **indented by one more space** to keep the triangle centered.

### Visual goal

For `draw_triangle(5)`, the output should look like:

```
 *****
  ***
   *
```

For `draw_triangle(9)`, the output should look like:

```
 *********
  *******
   *****
    ***
     *
```

Each row:
- Starts with **one more leading space** than the row above
- Has **two fewer stars** than the row above
- Ends when we reach a single `*`

The challenge is to write this using **recursion** — meaning the function calls itself to handle the smaller triangle below the current row.

---
## Section 2: First Experiment — Draw the Shape Manually

Before writing any recursive code, it helps to **understand the output target** exactly.

Let's manually print a triangle with base 5, hard-coding every row. This is **not** recursive — we are just getting a clear picture of the shape we want to produce.

Pay attention to:
- How many leading spaces each row has
- How many stars each row has
- The relationship between them

````python
# Manual triangle for base = 5
# This is NOT recursive. We are just printing each row by hand.

print(' *****')   # row 1: 1 space, 5 stars
print('  ***')    # row 2: 2 spaces, 3 stars
print('   *')     # row 3: 3 spaces, 1 star
````

### What do we notice?

Looking at the output above, we can see a clear pattern:

| Row | Leading spaces | Number of stars |
|-----|---------------|-----------------|
| 1   | 1             | 5               |
| 2   | 2             | 3               |
| 3   | 3             | 1               |

The **number of stars decreases by 2** each row, and the **number of leading spaces increases by 1** each row.

This means we can describe any row using just two values:
- How many stars to print
- How many spaces to print before the stars

This is the key insight we will use to build the recursive function.

---
## Section 3: Thinking Recursively

### What does it mean to think recursively?

Recursion is when a function **calls itself** to solve a smaller version of the same problem.

Instead of thinking "I need to draw 5 rows", recursion asks us to think:

> "To draw a triangle with base 5, I first print the top row (5 stars), and then I draw a **smaller triangle with base 3** directly below it."

And to draw a triangle with base 3:

> "I print the top row (3 stars), and then I draw a **smaller triangle with base 1** below it."

And to draw a triangle with base 1:

> "I print the single star. There is nothing left to draw."

### The recursive pattern

```
draw_triangle(5)
    print row with 5 stars
    draw_triangle(3)           <-- same problem, but smaller
        print row with 3 stars
        draw_triangle(1)       <-- smaller again
            print row with 1 star
            STOP               <-- can't go smaller
```

This is the essence of recursion:
1. Do something for the **current step**
2. Call yourself to handle the **rest of the problem** (which is slightly smaller)
3. Know when to **stop**

Notice that the triangle with base 3 is just the bottom portion of the triangle with base 5. Each recursive call handles one level of the triangle.

---
## Section 4: The Base Case — When to Stop

Every recursive function **must** have a **base case**: a situation where the function stops calling itself and simply returns.

Without a base case, the function would call itself forever, eventually crashing with a `RecursionError`.

For our triangle:
- The triangle keeps shrinking: 5 → 3 → 1
- When the base reaches **1**, we print a single star and stop
- There is no triangle with 0 stars or -1 stars to draw

Let's look at a small function fragment that **only** handles the base case, so we can see it clearly in isolation.

````python
# Base case demonstration
# This function only handles the smallest meaningful triangle (base = 1)

def draw_triangle_base_only(n, spaces=0):
    if n <= 0:
        # Base case: nothing to draw
        return
    
    # Print the current row
    print(' ' * spaces + '*' * n)
    
    # NOTE: We are NOT recursing yet. 
    # For now, we just print the first row and stop.

# Test it
print('draw_triangle_base_only(1):')
draw_triangle_base_only(1)

print()
print('draw_triangle_base_only(5):')
draw_triangle_base_only(5)  # Only prints the top row — recursion is missing
````

### What we see

- When `n = 1`, the function prints one star and stops. That is the base case working correctly.
- When `n = 5`, the function only prints the top row because we have not added the recursive call yet.

The key lesson: **checking `n <= 0` prevents the function from recursing infinitely.**

We use `n <= 0` (rather than just `n == 0`) as a safety net. Since we always reduce by 2 and start from an odd number, we go 5 → 3 → 1 and never hit 0 — but it's good defensive practice to guard against unexpected values.

Now let's add the recursive call.

---
## Section 5: Building the Recursive Function — Stage by Stage

We will build `draw_triangle()` in three stages:

1. **Stage 1** — Print stars only, no indentation
2. **Stage 2** — Add indentation (leading spaces)
3. **Stage 3** — Final polished version

Building in stages helps us focus on one problem at a time.

---

### Stage 1: Stars only (no spaces yet)

First, let's get the recursion working without worrying about formatting. We will print the correct number of stars on each row and shrink by 2 each call.

````python
# Stage 1: Recursion with stars only — no indentation yet

def draw_triangle_v1(n):
    # Base case: if n is 0 or negative, there's nothing to draw
    if n <= 0:
        return
    
    # Print the current row: n stars
    print('*' * n)
    
    # Recursive case: draw the next (smaller) row
    # We reduce n by 2 to shrink the triangle
    draw_triangle_v1(n - 2)

print('draw_triangle_v1(5):')
draw_triangle_v1(5)

print()
print('draw_triangle_v1(9):')
draw_triangle_v1(9)

print()
print('draw_triangle_v1(15):')
draw_triangle_v1(15)
````

### What Stage 1 shows us

The recursion is working! The triangle is shrinking correctly: 5 → 3 → 1, then stopping.

The only problem is that all rows are left-aligned. The original problem wants the triangle to look like it's centered, with increasing indentation as you go down.

**Key observation:** Each time we recurse, we need **one more leading space** than the previous call. That means indentation grows as `n` shrinks.

How do we pass that growing indentation to each recursive call? We add a second parameter.

---

### Stage 2: Adding indentation

We introduce a `spaces` parameter. It starts at 0 for the first call (no indentation on the top row), and increases by 1 with every recursive call.

````python
# Stage 2: Add a 'spaces' parameter to track indentation

def draw_triangle_v2(n, spaces):
    # Base case
    if n <= 0:
        return
    
    # Print spaces first, then stars
    print(' ' * spaces + '*' * n)
    
    # Recursive case:
    #   - n shrinks by 2 (fewer stars)
    #   - spaces grows by 1 (more indentation)
    draw_triangle_v2(n - 2, spaces + 1)

print('draw_triangle_v2(5, 1):')
draw_triangle_v2(5, 1)   # We must pass the starting spaces manually for now

print()
print('draw_triangle_v2(9, 1):')
draw_triangle_v2(9, 1)

print()
print('draw_triangle_v2(15, 1):')
draw_triangle_v2(15, 1)

print()
print('draw_triangle_v2(35, 1):')
draw_triangle_v2(35, 1)
````

### What Stage 2 shows us

The triangle shape now looks correct! The indentation grows by one space per row, and the stars shrink by two per row.

There is one small awkwardness though: the caller has to remember to pass `1` as the starting spaces. It would be nicer if the function handled that automatically.

That is exactly what **default parameter values** are for in Python. We can set `spaces=1` as the default, so callers just write `draw_triangle(5)` without worrying about the second argument.

Let's put together the final polished version.

---
## Section 6: The Final Recursive Solution

Now let's write the clean, final version of `draw_triangle()`. This version:
- Has a default value of `spaces=1` so callers don't need to pass it
- Has a clear base case
- Passes the updated values through each recursive call
- Produces correct output for any valid odd input less than 20

Read through the function and the comments carefully before running it.

````python
# Final version of draw_triangle()

def draw_triangle(n, spaces=1):
    """
    Recursively draws an upside-down isosceles triangle.
    
    Parameters:
        n      -- the number of stars on the current row (must be odd, < 20)
        spaces -- the number of leading spaces before the stars (default: 1)
    """
    # Base case: if there are no stars to print, we are done
    if n <= 0:
        return
    
    # Print the current row: leading spaces followed by stars
    print(' ' * spaces + '*' * n)
    
    # Recursive case:
    #   - The next row has 2 fewer stars
    #   - The next row has 1 more leading space
    draw_triangle(n - 2, spaces + 1)


# Quick test
print('draw_triangle(5):')
draw_triangle(5)
````

### Breaking down the final function

Let's look at each piece:

**`def draw_triangle(n, spaces=1):`**  
`n` is the current number of stars. `spaces` has a **default value** of 1, which means callers can just write `draw_triangle(5)` and the first row will automatically start with one leading space.

**`if n <= 0: return`**  
This is the **base case**. When there are no more stars to draw, the function returns without doing anything. This is what stops the recursion.

**`print(' ' * spaces + '*' * n)`**  
This prints the current row. `' ' * spaces` creates a string of spaces, and `'*' * n` creates a string of stars. The `+` joins them.

**`draw_triangle(n - 2, spaces + 1)`**  
This is the **recursive call**. Notice two things:
- `n - 2` shrinks the problem (fewer stars on the next row)
- `spaces + 1` passes the updated indentation to the next call

The function does not loop — it calls itself with a smaller problem, and that process repeats until `n <= 0`.

---
## Section 7: Tracing Through the Recursive Calls

One of the best ways to understand recursion is to **trace what happens** when you call a function.

Let's trace `draw_triangle(5)` step by step.

Each time the function calls itself, Python puts the current call on hold and starts a new one. This structure is called the **call stack**.

Here's what happens:

```
draw_triangle(5, spaces=1)
    prints: ' *****'           (1 space + 5 stars)
    calls: draw_triangle(3, spaces=2)
        prints: '  ***'        (2 spaces + 3 stars)
        calls: draw_triangle(1, spaces=3)
            prints: '   *'     (3 spaces + 1 star)
            calls: draw_triangle(-1, spaces=4)
                n <= 0 → returns immediately  ← BASE CASE REACHED
            draw_triangle(1) is done
        draw_triangle(3) is done
    draw_triangle(5) is done
```

Notice that each call prints **its own row**, then delegates the rest to the next call.

Let's write code to visualize this trace in the notebook.

````python
# Tracing version: prints diagnostic output so we can see the call stack

def draw_triangle_traced(n, spaces=1, depth=0):
    indent = '  ' * depth  # Indent the trace output based on recursion depth
    
    print(f"{indent}draw_triangle_traced({n}, spaces={spaces}) called")
    
    if n <= 0:
        print(f"{indent}  --> Base case reached (n={n}), returning")
        return
    
    row = ' ' * spaces + '*' * n
    print(f"{indent}  --> Printing row: '{row}'")
    print(row)  # Actually print the row
    
    print(f"{indent}  --> Recursing with n={n-2}, spaces={spaces+1}")
    draw_triangle_traced(n - 2, spaces + 1, depth + 1)
    
    print(f"{indent}draw_triangle_traced({n}) done")


print('=== Traced call: draw_triangle_traced(5) ===')
print()
draw_triangle_traced(5)
````

### What the trace tells us

This traced version adds extra diagnostic print statements so you can see:
- Which function call is active at each moment
- What values of `n` and `spaces` each call is working with
- When the base case is reached
- That each call "returns" after the recursive call finishes

The indented trace output mirrors the **call stack**: deeper indentation means deeper recursion.

**Key takeaway:** Python keeps track of every call that hasn't finished yet. When the base case returns, execution "unwinds" back up the stack. For this problem, each call only prints and then delegates — so the unwinding is quiet (nothing happens on the way back up). That's typical for many output-based recursive functions.

Once you understand this trace, you understand how recursion works.

## Section 9: Common Mistakes to Avoid

When students first write recursive functions, certain mistakes come up again and again. Here are the most common ones for this problem — and how to avoid them.

---

### Mistake 1: Forgetting the base case

```python
def draw_triangle_broken(n, spaces=1):
    print(' ' * spaces + '*' * n)
    draw_triangle_broken(n - 2, spaces + 1)  # Calls forever!
```

Without `if n <= 0: return`, the function will keep calling itself with -1, -3, -5, ... until Python raises a `RecursionError: maximum recursion depth exceeded`.

**Fix:** Always write your base case first.

---

### Mistake 2: Reducing the problem incorrectly

```python
draw_triangle(n - 1, spaces + 1)  # Wrong! Should be n - 2
```

If we subtract 1 instead of 2, the triangle won't have the right shape. The base shrinks by 2 each row (odd → odd: 5, 3, 1), not by 1.

**Fix:** Understand the pattern before coding it. We showed the table in Section 2 — the stars decrease by 2 per row.

---

### Mistake 3: Printing in the wrong order

```python
def draw_triangle_wrong_order(n, spaces=1):
    if n <= 0:
        return
    draw_triangle_wrong_order(n - 2, spaces + 1)  # Recurse first...
    print(' ' * spaces + '*' * n)                 # ...then print
```

This prints the triangle **upside up** — the smallest row prints first because the recursion runs all the way down before printing anything.

**Fix:** Print the current row **before** making the recursive call.

---

### Mistake 4: Mishandling spaces

```python
print('*' * n)             # Forgot the leading spaces
print(' ' * n + '*' * n)   # Wrong number of spaces (uses n instead of spaces)
```

The spacing is what makes the triangle look like it's centered. Missing or incorrect spaces produce a jagged or left-aligned shape.

**Fix:** Use the `spaces` parameter — that's exactly what it's for. `' ' * spaces + '*' * n` is the correct row.

---

### Mistake 5: Confusing parameters

A common confusion: students sometimes think `spaces` should shrink as `n` shrinks, because "there are fewer stars". But the opposite is true — spaces **grow** as stars shrink, because we're indenting more with each row.

**Fix:** Go back to the table in Section 2 and trace through by hand. The relationship is: `spaces` increases by 1, `n` decreases by 2, every row.

## Section 10: Recursion vs. Iteration — A Brief Comparison

Recursion is not the only way to solve this problem. A programmer thinking iteratively might reach for a `while` loop instead. Let's look at both side by side.

---

### The iterative approach

```python
def draw_triangle_iterative(n):
    spaces = 1
    while n > 0:
        print(' ' * spaces + '*' * n)
        n -= 2
        spaces += 1
```

This loop does the same thing as our recursive function:
- Print the current row
- Shrink `n` by 2
- Increase `spaces` by 1
- Repeat until `n <= 0`

---

### How are they different?

| | Recursive | Iterative |
|---|---|---|
| **State tracking** | Via function parameters passed forward | Via variables that change in the loop |
| **Stopping condition** | Base case (`if n <= 0: return`) | Loop condition (`while n > 0`) |
| **Readability** | Expresses "what" to do at each level | Expresses "how" to step through |
| **Complexity** | Same here; can be simpler for tree-like problems | Often simpler for linear sequences |

For this triangle problem, both solutions are equally valid. The iterative version might feel more natural if you're used to loops.

---

### Why learn recursion then?

Recursion becomes **essential** for problems where a loop would be very hard to write:
- Navigating folder structures (folders inside folders inside folders...)
- Traversing trees and graphs
- Solving puzzles like the Tower of Hanoi
- Implementing algorithms like merge sort or quicksort

For those problems, the recursive way of thinking — "solve this level, then let the function handle the rest" — is much clearer than any loop you could write.

The triangle problem is a simple, visual entry point to that way of thinking.

````python
# Iterative version for comparison

def draw_triangle_iterative(n):
    spaces = 1
    while n > 0:
        print(' ' * spaces + '*' * n)
        n -= 2
        spaces += 1

print('Iterative version — draw_triangle_iterative(9):')
draw_triangle_iterative(9)

print()
print('Recursive version — draw_triangle(9):')
draw_triangle(9)
````

The outputs are identical. The difference is how each function thinks about the problem.

The iterative function says: *"Keep going until I run out of stars."*  
The recursive function says: *"Print this row, then draw a smaller triangle."*

Both get to the same place.

## Section 11: Summary and Key Takeaways

### What is recursion?

Recursion is when a function **calls itself** to solve a smaller version of the same problem. Each call handles one piece of the problem, and the rest is delegated to the next call.

### How did this triangle problem demonstrate recursion?

- The triangle can be described as: **one row** + a **smaller triangle below it**
- That "smaller triangle below it" is the same problem with a smaller `n`
- The recursion stops when `n` reaches 0 or less (the base case)
- Each call passes updated values forward: `n - 2` (fewer stars) and `spaces + 1` (more indentation)

### The three rules of a recursive function

1. **Base case** — There must be a condition that stops the recursion
2. **Recursive case** — The function must call itself with a problem that is provably smaller
3. **Progress** — Each call must move closer to the base case (otherwise it runs forever)

### What to remember going forward

- Always write the **base case first** before you write the recursive call
- Ask yourself: *"What is the smallest version of this problem?"* — that's your base case
- Ask yourself: *"If I solve one step, what smaller problem is left?"* — that's your recursive call
- Trace through small examples by hand to verify your logic before testing on larger inputs
- Recursion and iteration are two different ways to think — both are valid tools, and knowing both makes you a stronger programmer
