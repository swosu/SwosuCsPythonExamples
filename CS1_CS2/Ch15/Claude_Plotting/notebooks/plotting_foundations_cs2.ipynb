## Part 1 - Why Plot?

Raw numbers are manageable when there are only a few of them, but long tables become slow to read and hard to compare. A plot turns those numbers into a picture so your brain can spot patterns in seconds instead of hunting for them row by row.

Programmers use plots to compare algorithms, debug unexpected behavior, and communicate results to other people. A plot is not decoration. A plot is evidence.

### What Makes a Good Plot?

Every good plot begins with a question. If you do not know what you are trying to answer, the chart will usually feel confusing.

A strong plot also needs axis labels so the reader knows what the numbers mean, a title that points to the main idea, and a legend whenever more than one data series appears.

A grid can help readability, but too much visual clutter hurts. The best plot is not the fanciest plot - it is the one that answers the question clearly.

### What Makes a Plot Go Wrong?

Before we write a single line of `matplotlib`, we need to look at what bad plots look like and why they fail. A plot can be technically correct and still mislead the reader. Common mistakes happen before you type a single line of code.

1. **No title.** The reader has no idea what question the plot is answering.
2. **No axis labels.** Numbers without units or context are meaningless. Is the y axis dollars? Students? Seconds?
3. **Missing legend with multiple series.** Two lines with no labels are indistinguishable.
4. **Y-axis that doesn't start at zero (when it should).** A bar chart starting at 85 instead of 0 makes a small difference look enormous.
5. **Too much clutter.** Gridlines, borders, tick marks, and colors fighting each other distract from the data.
6. **Wrong chart type.** Using a line chart for unordered categories makes the trend look meaningful when it is not.

We will now produce a deliberately bad plot and then fix it. Both versions use the same data. The difference is entirely in the choices the programmer made.

````python
# This is a deliberately bad plot.
# It commits three mistakes at once: no title, no labels, truncated y-axis.
scores = [88, 91, 79, 85, 94, 77, 83, 90]
plt.figure(figsize=(5, 3))
plt.bar(range(len(scores)), scores)
plt.ylim(75, 96)
plt.show()
````

What is wrong with this plot? List at least three problems before reading on.

- No title: we cannot tell what these scores represent.
- No x-axis label: what are the bars? Students? Assignments? Days?
- No y-axis label: are these points, percentages, dollars?
- The y-axis starts at 75, not 0. The difference between 77 and 94 looks like a cliff, not a gap of 17 points. This is a common misleading chart technique.

````python
# The same data, fixed.
scores = [88, 91, 79, 85, 94, 77, 83, 90]
student_labels = [f"S{i+1}" for i in range(len(scores))]

plt.figure(figsize=(8, 5))
plt.bar(student_labels, scores, color='steelblue', edgecolor='black')
plt.title("Quiz Scores by Student (Synthetic Data)")
plt.xlabel("Student")
plt.ylabel("Score (out of 100)")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig("figures/bad_vs_better_plot.png")
plt.show()

print("Figure saved to figures/bad_vs_better_plot.png")
````

- Title added: the reader knows what this is.
- Axis labels added: y-axis says `Score (out of 100)` so there is no ambiguity.
- Y-axis starts at 0: a score of 77 looks like 77% of the bar, not like a failure.
- Grid on y-axis only: this helps read values without adding horizontal clutter.
- Labels on x-axis: `S1` through `S8` are generic, but they exist.

> **The rule:** A plot should answer one question clearly. Every element - title, labels, scale, color, legend - should serve that question. Remove anything that does not.

| If your x axis is... | And you want to show... | Use... |
|---|---|---|
| Categories (unordered) | Comparison across groups | Bar chart |
| Time or sequence | Trend over time | Line chart |
| Two numeric variables | Relationship / correlation | Scatter plot |
| Same numeric variable, two groups | Side-by-side comparison | Subplot or grouped bar |

When in doubt, sketch it on paper first. If the sketch doesn't answer the question, the code won't either.

## Part 2 - Your First Line Plot

We will use two plain Python lists and `matplotlib` to draw a first line plot. No NumPy yet, no file loading, and no extra steps beyond connecting values on the x-axis to values on the y-axis.

````python
import matplotlib.pyplot as plt
````

We need two lists: one for x values on the horizontal axis and one for y values on the vertical axis. In this example, the x values are weeks and the y values are fictional temperature readings.

````python
weeks = [1, 2, 3, 4, 5, 6, 7, 8]
avg_temp = [58, 62, 67, 71, 75, 73, 69, 64]

plt.plot(weeks, avg_temp)
plt.title("Average Weekly Temperature")
plt.xlabel("Week")
plt.ylabel("Temperature (F)")
plt.grid(True)
plt.show()
````

- `plt.plot(weeks, avg_temp)` - draws a line connecting each `(week, temp)` pair.
- `plt.title(...)` - sets the chart title.
- `plt.xlabel(...)` and `plt.ylabel(...)` - label the axes.
- `plt.grid(True)` - adds a light grid to help read values.
- `plt.show()` - renders the figure.

Now we add a legend. Legends matter when more than one line appears, because the reader needs a reliable way to tell the series apart.

````python
weeks = [1, 2, 3, 4, 5, 6, 7, 8]
city_a = [58, 62, 67, 71, 75, 73, 69, 64]
city_b = [45, 50, 55, 60, 63, 61, 57, 52]

plt.plot(weeks, city_a, label="City A")
plt.plot(weeks, city_b, label="City B")
plt.title("Weekly Temperature Comparison")
plt.xlabel("Week")
plt.ylabel("Temperature (F)")
plt.legend()
plt.grid(True)
plt.savefig("figures/basic_line_plot.png")
plt.show()
print("Figure saved to figures/basic_line_plot.png")
````

Legends are required when more than one series appears. Without a legend, the reader can see two lines but cannot confidently tell which one belongs to which dataset. `plt.savefig()` must come before `plt.show()` because after `show()`, the figure may be cleared. Saving a plot creates a file that proves the code ran. That file is your run result.

## Part 3 - NumPy Arrays as Plot Data

NumPy arrays work like lists for plotting, but they also give us built-in math operations such as `.mean()`, `.median()`, and `.max()`. In this section, we will use a synthetic set of exam scores to summarize and visualize the data.

````python
import numpy as np
````

We create a synthetic array of exam scores so we control exactly what we are working with. In a real project, this data might come from a file, but hard-coded values keep the lesson focused on the plotting ideas.

````python
exam_scores = np.array([72, 85, 91, 68, 77, 88, 95, 61, 74, 83, 79, 90, 55, 82, 78])

average = np.mean(exam_scores)
median  = np.median(exam_scores)
highest = np.max(exam_scores)
count_above_80 = np.sum(exam_scores >= 80)

print(f"Average score : {average:.1f}")
print(f"Median score  : {median:.1f}")
print(f"Highest score : {highest}")
print(f"Scores >= 80  : {count_above_80}")
````

The first line creates a NumPy array of scores. The next three lines compute the average, median, and maximum score. `np.sum(exam_scores >= 80)` counts how many scores meet the threshold because the comparison creates a Boolean array and `True` counts as `1`.

Now we plot each score against a student index. We also add a horizontal reference line at `80` so it is easy to see who is above or below that threshold.

````python
student_numbers = np.arange(1, len(exam_scores) + 1)

plt.figure(figsize=(10, 5))
plt.plot(student_numbers, exam_scores, marker='o', label="Exam Score")
plt.axhline(y=80, color='red', linestyle='--', label="Score = 80")
plt.title("Exam Scores by Student")
plt.xlabel("Student Number")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.savefig("figures/exam_average_plot.png")
plt.show()
print("Figure saved to figures/exam_average_plot.png")
````

`plt.axhline()` draws a horizontal reference line across the plot. Reference lines are useful when there is a threshold, cutoff, or target value, because they turn a raw number into a visible comparison point. Every plot in this notebook is saved to `figures/` with a descriptive filename. This is the pattern for any assignment where you need to submit proof of output: write the code, save the figure, then use the saved file as your run result.

## Part 4 - Transforming Data: Score Curving

Array transformation means applying the same operation to every element at once. NumPy makes that kind of change concise and readable. In this example, we will curve scores by shifting the whole array upward so the maximum score becomes `100`.

The formula is `curved = original_scores + (100 - max_score)`. Because the maximum score shifts to `100`, every other score shifts by exactly the same amount.

````python
original_scores = np.array([72, 85, 88, 68, 77, 81, 76, 61, 74, 83, 79, 82, 55, 80, 78])
max_score = np.max(original_scores)
curved_scores = original_scores + (100 - max_score)

print(f"Max original : {max_score}")
print(f"Curve added  : {100 - max_score} points")
print(f"Max curved   : {np.max(curved_scores)}")
print(f"\nOriginal : {original_scores}")
print(f"Curved   : {curved_scores}")
````

NumPy broadcasts the scalar addition across the entire array, so there is no loop here. That is one of the biggest reasons NumPy is useful for data work: one clear expression can transform every value at once.

Now we will plot both arrays on the same chart so the transformation becomes visible. The overall shape should stay the same, but every value should be higher after curving.

````python
student_numbers = np.arange(1, len(original_scores) + 1)

plt.figure(figsize=(10, 5))
plt.plot(student_numbers, original_scores, marker='o', label="Original Scores")
plt.plot(student_numbers, curved_scores, marker='s', linestyle='--', label="Curved Scores")
plt.title("Score Distribution Before and After Curving")
plt.xlabel("Student Number")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.savefig("figures/curved_scores_plot.png")
plt.show()

print("Figure saved to figures/curved_scores_plot.png")
````

Plotting transformations helps verify that the math worked the way you expected. The two lines should be parallel because the transformation adds the same amount to every score. If they crossed, something would be wrong. `plt.savefig()` comes before `plt.show()` so the figure is written to disk before the display step can clear it.

## Part 5 - Pandas DataFrames and Bar Charts

Pandas DataFrames are a common way to store tabular data in Python. Each row is an observation, and each column is a variable. Here we will compute summary statistics and build a bar chart showing the class average per assignment.

````python
import pandas as pd
````

We create the DataFrame directly from a dictionary of lists. In a real project, this might come from a CSV file. Here we hard-code it so the lesson stays focused on analysis and plotting.

````python
data = {
    "Student": ["Student A", "Student B", "Student C", "Student D",
                "Student E", "Student F", "Student G", "Student H"],
    "Assignment 1": [88, 75, 92, 61, 79, 85, 70, 95],
    "Assignment 2": [82, 68, 88, 74, 80, 91, 65, 90],
    "Assignment 3": [90, 72, 95, 70, 85, 88, 77, 93],
}

df = pd.DataFrame(data)
df_sorted = df.sort_values("Assignment 3", ascending=False)
print(df_sorted.to_string(index=False))
````

`.sort_values()` reorders rows by a column. `ascending=False` puts the highest score first. `to_string(index=False)` prints a clean table without the row index numbers.

Now we compute summary statistics across the numeric columns.

````python
numeric_cols = ["Assignment 1", "Assignment 2", "Assignment 3"]

print("Class Statistics")
print(f"  Max    : {df[numeric_cols].max(numeric_only=True).values}")
print(f"  Median : {df[numeric_cols].median(numeric_only=True).values}")
print(f"  Mean   : {df[numeric_cols].mean(numeric_only=True).values.round(1)}")
print(f"  Std Dev: {df[numeric_cols].std(numeric_only=True).values.round(1)}")
````

Bar charts work when the x axis is categorical, not numeric. `plt.ylim(0, 100)` anchors the scale so small differences do not look larger than they are, and `grid(axis='y')` adds horizontal gridlines only for a cleaner look.

## Part 7 - Scatter Plot Story: Capacity vs. Potential

Scatter plots are the right tool when you have two numeric variables and you want to see whether they are related. We will use synthetic venue data: seating capacity on the x axis and gross revenue potential on the y axis.

We create a small synthetic DataFrame. We also add a `Marker Size` column so larger venues appear as larger dots, a technique called encoding a third variable with visual size.

````python
venue_data = pd.DataFrame({
    "Venue": ["Majestic", "Palace", "Roxy", "Orpheum", "Apollo",
              "Grand", "Bijou", "Forum", "Strand", "Imperial"],
    "Capacity": [1800, 2200, 950, 1400, 1100, 3000, 750, 2600, 1250, 1700],
    "Gross Potential": [95000, 130000, 42000, 70000, 55000,
                        175000, 30000, 160000, 62000, 88000],
})

venue_data["Marker Size"] = venue_data["Gross Potential"] / 1000

print(venue_data[["Venue", "Capacity", "Gross Potential"]].to_string(index=False))
````

`plt.scatter()` works like `plt.plot()`, but it draws individual points instead of connecting them. We pass the `s` argument for marker size and can also pass `c` for color.

````python
plt.figure(figsize=(9, 6))
plt.scatter(
    venue_data["Capacity"],
    venue_data["Gross Potential"],
    s=venue_data["Marker Size"],
    c='steelblue',
    alpha=0.7,
    edgecolors='black'
)

plt.title("Venue Capacity vs. Gross Revenue Potential (Synthetic Data)")
plt.xlabel("Seating Capacity")
plt.ylabel("Gross Potential ($)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("figures/capacity_gross_scatter.png")
plt.show()

print("Figure saved to figures/capacity_gross_scatter.png")
````

- `s=` controls the marker size. Here we scale it from gross potential.
- `alpha=0.7` makes markers slightly transparent so overlapping points are visible.
- `edgecolors='black'` adds a thin border to each marker for readability.
- `linestyle='--'` on the grid makes it softer than the solid default.

Scatter plots reveal relationships that tables hide. A trend line could be added later to summarize the pattern. Encoding a third variable in marker size is common, but the size scale should be labeled if the reader needs exact values.

## Part 8 - Subplots: Comparing Two Datasets

Sometimes you want to compare two plots side by side. `plt.subplots()` creates a figure with multiple panels, called axes. Each panel can hold its own plot.

We use two synthetic monthly datasets for fictional product lines Alpha and Beta.

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

sales_alpha = [420, 390, 450, 480, 510, 540, 600, 590, 530, 470, 440, 620]
sales_beta  = [310, 340, 370, 320, 360, 410, 450, 430, 390, 350, 380, 470]
````

`plt.subplots(1, 2)` creates one row and two columns of panels. The function returns the figure object and an array of axes objects. We plot into each axis separately.

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Monthly Sales Comparison: Alpha vs. Beta (Synthetic Data)", fontsize=14)

axes[0].scatter(months, sales_alpha, color='steelblue', edgecolors='black')
axes[0].set_title("Product Line Alpha")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Units Sold")
axes[0].grid(True, linestyle='--', alpha=0.5)

axes[1].scatter(months, sales_beta, color='darkorange', edgecolors='black')
axes[1].set_title("Product Line Beta")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Units Sold")
axes[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("figures/broadway_style_subplots.png")
plt.savefig("subplots.png")
plt.show()

print("Figures saved.")
````

- `fig.suptitle()` sets a title for the whole figure, not just one panel.
- `axes[0]` and `axes[1]` refer to the left and right panels.
- When using subplots, use `axes[i].set_title()`, `axes[i].set_xlabel()`, and `axes[i].set_ylabel()` instead of the plain `plt` versions.
- We save to both a descriptive filename and `subplots.png` for lab compatibility.

Side-by-side subplots force the axes to share the same visual space, making trends easier to compare. If the y-axis scales differ significantly, that can mislead a reader, so matching scales may be important when comparison is the goal.

## Part 9 - Fibonacci Algorithm Story

Fibonacci is a great algorithm story because iterative and recursive solutions produce the same correct answers but behave very differently in cost. Plotting makes the difference impossible to ignore.

Recall the sequence: 0, 1, 1, 2, 3, 5, 8, 13, 21...

The iterative version computes Fibonacci using a loop. It remembers only the two most recent values and replaces them each step.

````python
def fib_iterative(n):
    if n <= 0:
        return 0
    if n == 1:
        return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b
````

This function does exactly `n - 1` additions. It runs in O(n) time and uses O(1) space because it keeps only two variables regardless of `n`.

The recursive version is closer to the mathematical definition. It calls itself twice for each step, which sounds harmless at first.

````python
call_count = 0

def fib_recursive(n):
    global call_count
    call_count += 1
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fib_recursive(n - 1) + fib_recursive(n - 2)
````

`global call_count` lets us count how many times the function is called. We will reset it before each measurement. The function produces the correct answer but at potentially enormous cost.

Let us verify both functions return the same values for small `n`, and see the call count for the recursive version.

````python
print("n  | iterative | recursive | calls")
print("---|-----------|-----------|------")

for n in range(0, 16):
    call_count = 0
    result_iter = fib_iterative(n)
    result_rec = fib_recursive(n)
    print(f"{n:2d} | {result_iter:9d} | {result_rec:9d} | {call_count:6d}")
````

Notice how the call count grows. For `n=10` it is already in the hundreds. For `n=20` it will be in the thousands. For `n=35` it will be in the millions. That growth is what we are about to measure and plot.

### Warning

We will cap naive recursive Fibonacci at `n=28` in this notebook. Running the recursive version for `n=40+` can freeze a laptop. The iterative version can safely run for much larger `n`.

We will now collect two measurements for each safe `n` value: wall-clock time in seconds using `time.perf_counter()` and function call count using our global counter. Then we will plot both metrics.

````python
import time
````

We build lists by looping over our `n` values and recording each measurement. For recursive Fibonacci we reset `call_count` before each call so every measurement is independent.

````python
n_values_iter = list(range(0, 41))
n_values_rec = list(range(0, 29))

iter_times = []
rec_times = []
rec_calls = []
iter_calls = []

for n in n_values_iter:
    start = time.perf_counter()
    fib_iterative(n)
    iter_times.append(time.perf_counter() - start)
    iter_calls.append(n)

for n in n_values_rec:
    call_count = 0
    start = time.perf_counter()
    fib_recursive(n)
    elapsed = time.perf_counter() - start
    rec_times.append(elapsed)
    rec_calls.append(call_count)

print(f"Iterative measured for n = 0 to {n_values_iter[-1]}")
print(f"Recursive measured for n = 0 to {n_values_rec[-1]}")
print("\nSample recursive call counts:")
for i in [0, 5, 10, 15, 20, 25, 28]:
    print(f"  n={i:2d}  calls={rec_calls[i]:>8,}")
````

`time.perf_counter()` gives high-resolution elapsed time in seconds. Timing a single call at small `n` is noisy and may be near zero, so call counts are often a cleaner signal. We collect both so students can compare.

Wall-clock timing is affected by what else is running on the computer. A single call to `fib_iterative(5)` might show slightly different values on different runs. Call counts do not have this problem because they are exact.

We now have our data. We will plot call counts and timing on separate panels so students can see both stories at once.

We use a two-panel subplot: the left panel shows call counts and the right panel shows wall-clock time. Both compare iterative growth against naive recursive growth.

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Iterative vs. Naive Recursive Fibonacci: Growth Comparison", fontsize=14)

axes[0].plot(n_values_iter, n_values_iter,
             label="Iterative (O(n) operations)", color='steelblue', linewidth=2)
axes[0].plot(n_values_rec, rec_calls,
             label="Recursive (exponential calls)", color='crimson', linewidth=2)
axes[0].set_title("Function Call Counts")
axes[0].set_xlabel("n")
axes[0].set_ylabel("Number of Calls / Operations")
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.5)

axes[1].plot(n_values_iter, iter_times,
             label="Iterative", color='steelblue', linewidth=2)
axes[1].plot(n_values_rec, rec_times,
             label="Recursive", color='crimson', linewidth=2)
axes[1].set_title("Wall-Clock Time (seconds)")
axes[1].set_xlabel("n")
axes[1].set_ylabel("Time (seconds)")
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("figures/fibonacci_growth.png")
plt.show()

print("Figure saved to figures/fibonacci_growth.png")
````

The left panel makes exponential growth visible. The recursive line bends upward sharply while the iterative line stays linear. The right panel is noisier because timing varies with system load, but the trend is the same. This is why computer scientists care about time complexity: a correct answer that takes exponential time can become practically unusable.

Naive recursive Fibonacci computes the same sub-problems repeatedly. To find `fib(10)`, it computes `fib(9)` and `fib(8)`. To find `fib(9)`, it computes `fib(8)` again. Every level of the call tree branches twice, so the total work grows at roughly `2^n`.

```
fib(5)
|- fib(4)
|  |- fib(3)
|  \- fib(2)
\- fib(3)
   |- fib(2)
   \- fib(1)
```
