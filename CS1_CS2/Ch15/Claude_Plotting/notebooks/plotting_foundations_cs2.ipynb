## Part 1 - Why Plot?

Raw numbers are manageable when there are only a few of them, but long tables become slow to read and hard to compare. A plot turns those numbers into a picture so your brain can spot patterns in seconds instead of hunting for them row by row.

Programmers use plots to compare algorithms, debug unexpected behavior, and communicate results to other people. A plot is not decoration. A plot is evidence.

### What Makes a Good Plot?

Every good plot begins with a question. If you do not know what you are trying to answer, the chart will usually feel confusing.

A strong plot also needs axis labels so the reader knows what the numbers mean, a title that points to the main idea, and a legend whenever more than one data series appears.

A grid can help readability, but too much visual clutter hurts. The best plot is not the fanciest plot - it is the one that answers the question clearly.

## Part 2 - Your First Line Plot

We will use two plain Python lists and `matplotlib` to draw a first line plot. No NumPy yet, no file loading, and no extra steps beyond connecting values on the x-axis to values on the y-axis.

```python
import matplotlib.pyplot as plt
```

We need two lists: one for x values on the horizontal axis and one for y values on the vertical axis. In this example, the x values are weeks and the y values are fictional temperature readings.

```python
weeks = [1, 2, 3, 4, 5, 6, 7, 8]
avg_temp = [58, 62, 67, 71, 75, 73, 69, 64]

plt.plot(weeks, avg_temp)
plt.title("Average Weekly Temperature")
plt.xlabel("Week")
plt.ylabel("Temperature (F)")
plt.grid(True)
plt.show()
```

- `plt.plot(weeks, avg_temp)` - draws a line connecting each `(week, temp)` pair.
- `plt.title(...)` - sets the chart title.
- `plt.xlabel(...)` and `plt.ylabel(...)` - label the axes.
- `plt.grid(True)` - adds a light grid to help read values.
- `plt.show()` - renders the figure.

Now we add a legend. Legends matter when more than one line appears, because the reader needs a reliable way to tell the series apart.

```python
weeks = [1, 2, 3, 4, 5, 6, 7, 8]
city_a = [58, 62, 67, 71, 75, 73, 69, 64]
city_b = [45, 50, 55, 60, 63, 61, 57, 52]

plt.plot(weeks, city_a, label="City A")
plt.plot(weeks, city_b, label="City B")
plt.title("Weekly Temperature Comparison")
plt.xlabel("Week")
plt.ylabel("Temperature (F)")
plt.legend()
plt.grid(True)
plt.show()
```

Legends are required when more than one series appears. Without a legend, the reader can see two lines but cannot confidently tell which one belongs to which dataset. This is a very common student plotting mistake.

## Part 3 - NumPy Arrays as Plot Data

NumPy arrays work like lists for plotting, but they also give us built-in math operations such as `.mean()`, `.median()`, and `.max()`. In this section, we will use a synthetic set of exam scores to summarize and visualize the data.

```python
import numpy as np
```

We create a synthetic array of exam scores so we control exactly what we are working with. In a real project, this data might come from a file, but hard-coded values keep the lesson focused on the plotting ideas.

```python
exam_scores = np.array([72, 85, 91, 68, 77, 88, 95, 61, 74, 83, 79, 90, 55, 82, 78])

average = np.mean(exam_scores)
median  = np.median(exam_scores)
highest = np.max(exam_scores)
count_above_80 = np.sum(exam_scores >= 80)

print(f"Average score : {average:.1f}")
print(f"Median score  : {median:.1f}")
print(f"Highest score : {highest}")
print(f"Scores >= 80  : {count_above_80}")
```

The first line creates a NumPy array of scores. The next three lines compute the average, median, and maximum score. `np.sum(exam_scores >= 80)` counts how many scores meet the threshold because the comparison creates a Boolean array and `True` counts as `1`.

Now we plot each score against a student index. We also add a horizontal reference line at `80` so it is easy to see who is above or below that threshold.

```python
student_numbers = np.arange(1, len(exam_scores) + 1)

plt.figure(figsize=(10, 5))
plt.plot(student_numbers, exam_scores, marker='o', label="Exam Score")
plt.axhline(y=80, color='red', linestyle='--', label="Score = 80")
plt.title("Exam Scores by Student")
plt.xlabel("Student Number")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()
```

`plt.axhline()` draws a horizontal reference line across the plot. Reference lines are useful when there is a threshold, cutoff, or target value, because they turn a raw number into a visible comparison point.
