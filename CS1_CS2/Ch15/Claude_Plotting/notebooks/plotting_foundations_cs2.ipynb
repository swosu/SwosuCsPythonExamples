## Part 1 - Why Plot?

Raw numbers are manageable when there are only a few of them, but long tables become slow to read and hard to compare. A plot turns those numbers into a picture so your brain can spot patterns in seconds instead of hunting for them row by row.

Programmers use plots to compare algorithms, debug unexpected behavior, and communicate results to other people. A plot is not decoration. A plot is evidence.

### What Makes a Good Plot?

Every good plot begins with a question. If you do not know what you are trying to answer, the chart will usually feel confusing.

A strong plot also needs axis labels so the reader knows what the numbers mean, a title that points to the main idea, and a legend whenever more than one data series appears.

A grid can help readability, but too much visual clutter hurts. The best plot is not the fanciest plot - it is the one that answers the question clearly.
