Title: Understanding the Old NANP Telephone Number Format

The goal of this step is to restate the structure of the old North American Numbering Plan (NANP) and identify how many choices each digit position allows.
We will break the telephone number into:

Area code: NYX

Office code: NNX

Station code: XXXX

Where:

N = digits 2–9 (8 possibilities)

Y = digits 0–1 (2 possibilities)

X = digits 0–9 (10 possibilities)

We will compute the number of possible codes for each part.

```python
# Number of possibilities for each digit type
N = 8   # digits 2–9
Y = 2   # digits 0–1
X = 10  # digits 0–9

# Old plan formats
area_codes_old = N * Y * X
office_codes_old = N * N * X
station_codes = X ** 4

area_codes_old, office_codes_old, station_codes
```
