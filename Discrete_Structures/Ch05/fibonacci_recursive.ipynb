# Bonus: Fibonacci and duplicate work
The excerpt shows a recursion tree for Fibonacci evaluation and explains
how recursive Fibonacci repeats the same subproblems (pages 1–2).

We’ll count function calls to make the "duplicate work" visible.


```python
@dataclass
class FibStats:
    calls: int = 0

def fib_recursive(n, stats: FibStats):
    stats.calls += 1
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib_recursive(n-1, stats) + fib_recursive(n-2, stats)

def fib_iterative(n):
    if n == 0:
        return 0
    x, y = 0, 1
    for _ in range(1, n):
        x, y = y, x + y
    return y

for n in range(0, 16):
    s = FibStats()
    r = fib_recursive(n, s)
    i = fib_iterative(n)
    print(n, r, i, "calls:", s.calls)
```
