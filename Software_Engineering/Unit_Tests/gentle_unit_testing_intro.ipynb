# Gentle Introduction to Unit Testing

Unit testing matters because it gives us a quick way to check whether our code still behaves the way we expect. As programs get larger, it becomes harder to manually test every situation, so small automated tests help us catch mistakes early and build confidence.

In this notebook, we will move through a few examples with gently increasing complexity. The goal is not to make testing feel intimidating. The goal is to show that a unit test is just a small, focused check for one piece of code.

Run the notebook from top to bottom. Each code cell defines a small function, creates a few tests, and runs them right inside the notebook.

## Goal of Our 1st Example

Our first example has one very simple goal: test a function that adds two numbers. This lets us focus on the structure of a unit test without worrying about complicated logic.

```python
import unittest


def run_test_case(test_case):
    suite = unittest.defaultTestLoader.loadTestsFromTestCase(test_case)
    runner = unittest.TextTestRunner(verbosity=2)
    runner.run(suite)


def add_numbers(a, b):
    return a + b


class TestAddNumbers(unittest.TestCase):
    def test_adds_two_positive_numbers(self):
        self.assertEqual(add_numbers(2, 3), 5)

    def test_adds_a_negative_and_positive_number(self):
        self.assertEqual(add_numbers(-2, 5), 3)


run_test_case(TestAddNumbers)
```

### Why This Example Matters

This first example shows the basic shape of a unit test. We write a small function, create a class that inherits from `unittest.TestCase`, and then write methods whose names start with `test_`.

A few implementation details are worth noticing:

- `self.assertEqual(...)` checks that the actual result matches the expected result.
- Each test method should focus on one idea.
- `run_test_case(...)` lets the tests run neatly inside a notebook cell.

Even though this example is tiny, it teaches an important habit: when we write a function, we can also write a quick automated check for it.

## Goal of Our 2nd Example

Now we will test a function that works with strings instead of numbers. The goal is to show that unit tests are useful for checking formatting rules and multiple input cases.

```python
def format_full_name(first_name, last_name, middle_name=""):
    first_name = first_name.strip().capitalize()
    last_name = last_name.strip().capitalize()
    middle_name = middle_name.strip().capitalize()

    if middle_name:
        return f"{first_name} {middle_name} {last_name}"

    return f"{first_name} {last_name}"


class TestFormatFullName(unittest.TestCase):
    def test_formats_first_and_last_name(self):
        self.assertEqual(format_full_name("ada", "lovelace"), "Ada Lovelace")

    def test_formats_name_with_middle_name(self):
        self.assertEqual(
            format_full_name("grace", "hopper", "brewster"),
            "Grace Brewster Hopper"
        )

    def test_removes_extra_spaces(self):
        self.assertEqual(
            format_full_name("  alan", "turing  "),
            "Alan Turing"
        )


run_test_case(TestFormatFullName)
```

### Why This Example Matters

This example is a little richer because the function has more than one possible path. Sometimes there is a middle name, and sometimes there is not. Sometimes the input has extra spaces, and sometimes it does not.

That leads to an important testing idea: do not test only the most obvious case. Think about a few realistic variations.

A few implementation details stand out here:

- `.strip()` cleans up spaces at the beginning or end of a string.
- `.capitalize()` helps standardize the way the name is displayed.
- The three test methods each target a different behavior of the same function.

This is where unit testing starts to become especially helpful. It protects us when a function has small branches or formatting choices that are easy to break by accident.

## Goal of Our 3rd Example

Our third example adds one more idea: sometimes good code should reject bad input. We want to test both a correct answer and an error case.

```python
def split_bill(total_cost, number_of_people):
    if number_of_people <= 0:
        raise ValueError("number_of_people must be greater than 0")

    return total_cost / number_of_people


class TestSplitBill(unittest.TestCase):
    def test_splits_bill_evenly(self):
        self.assertAlmostEqual(split_bill(20, 4), 5.0)

    def test_handles_decimal_answer(self):
        self.assertAlmostEqual(split_bill(10, 3), 3.3333333333, places=3)

    def test_raises_error_for_zero_people(self):
        with self.assertRaises(ValueError):
            split_bill(20, 0)


run_test_case(TestSplitBill)
```

### Why This Example Matters

This example introduces two ideas that appear often in real software.

First, we sometimes work with decimal answers, so `assertAlmostEqual(...)` is useful when we care about values that may involve floating-point arithmetic.

Second, testing is not only about checking correct output. It is also about checking that our program responds properly to invalid input. Here, we use `with self.assertRaises(ValueError):` to confirm that bad input is rejected in a clear and intentional way.

That is an important shift in thinking. A good unit test suite checks both:

- what the code should do
- what the code should not allow

By this point, students can see that unit tests are really about describing expected behavior in a precise, repeatable way.
