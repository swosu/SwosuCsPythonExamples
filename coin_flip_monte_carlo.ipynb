# Monte Carlo Coin Flip Simulation with Rare Side Landing

## Overall Plan

We will simulate a coin that has three possible outcomes:

- Heads
- Tails
- Side (very rare)

Our goal is to determine:

1. How many flips are required to achieve:
   - 50% certainty
   - 75% certainty
   - 90% certainty
   - 95% certainty
   - 99% certainty

   that we will observe at least one "side" landing.

2. Validate those theoretical results using Monte Carlo simulation.

3. Estimate how many Monte Carlo simulation runs are required for stable estimates.

We will:
- Derive the formula mathematically
- Confirm it experimentally
- Visualize convergence

## Goal

Compute the number of flips required to reach specific certainty levels using the exact probability formula.

We solve:

1 - (1 - p_side)^n ≥ certainty

```python
import numpy as np
import pandas as pd

# Rare side probability
p_side = 1 / 6000

# Desired certainty levels
certainties = [0.50, 0.75, 0.90, 0.95, 0.99]

def required_flips(p, certainty):
    return np.log(1 - certainty) / np.log(1 - p)

results = []

for c in certainties:
    n = required_flips(p_side, c)
    results.append((c, int(np.ceil(n))))

df = pd.DataFrame(results, columns=["Certainty", "Required Flips"])
df
```
